```python
# find variance and standard devidation of an array
```

```python
import numpy as np
```

```python
myArray = np.random.normal(0, 1, 1000)
```

```python
np.mean(myArray)
```

```python
print(myArray[:10])
```

```python
np.var(myArray)
```

```python
np.std(myArray)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(myArray)
plt.show()
```
