```python
# how to use reduce in python
```

```python
import functools
```

```python
myList = [1, 2, 3, 4, 5]
```

```python
mySum = functools.reduce(lambda x, y: x + y, myList)
```

```python
print(mySum)
```

```python
myMul = functools.reduce(lambda x, y: x * y, myList)
```

```python
print(myMul)
```
