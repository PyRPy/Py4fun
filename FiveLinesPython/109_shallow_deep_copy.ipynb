```python
# what's shallow copy or deep copy in python
```

```python
myList = [1, 2, 3, 4, 5]
```

```python
shallowList = myList
```

```python
print(shallowList)
```

```python
# if we change original list
```

```python
myList[4] = 6
```

```python
print(myList)
```

```python
print(shallowList)
```

```python
shallowList[4] = 5
```

```python
print(myList)
```

```python
# deep copy
```

```python
import copy
```

```python
deepList = copy.deepcopy(myList)
```

```python
print(deepList)
```

```python
myList[4] = 6
```

```python
print(deepList)
```

```python
print(myList)
```
