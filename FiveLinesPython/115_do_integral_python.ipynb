```python
# how to do an integral in python
```

```python
from sympy import Integral, Symbol
```

```python
x = Symbol('x')
c = Symbol('c')
```

```python
Integral(c*x, x)
```

```python
Integral(c*x, x).doit()
```

```python
Integral(c*x, (x, 0, 1))
```

```python
Integral(c*x, (x, 0, 1)).doit()
```
