```python
# how to use decorators
```

```python
def action_on_pair(func):
  def inner_func(nested_list):
    return [func(val[0], val[1]) for val in nested_list]
  return inner_func
```

```python
@action_on_pair
def multiply_two(x, y):
  return x * y
```

```python
myList = [[1, 2], [3, 3], [5, 6]]
```

```python
print(multiply_two(myList))
```

```python
@action_on_pair
def compare_two(x, y):
  return x == y
```

```python
print(compare_two(myList))
```
