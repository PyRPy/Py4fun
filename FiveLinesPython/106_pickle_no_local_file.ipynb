```python
# what is 'pickle' in python
```

```python
student_id = [100, 101, 102, 103, 104]
student_name = ['David', 
                'Simon', 
                'Victor', 
                'Eric', 
                'Frank']
```

```python
students = zip(student_id, student_name)
```

```python
import pickle
```

```python
dat = pickle.dumps(students)
```

```python
type(dat)
```

```python
type(students)
```

```python
type(student_id)
```

```python
my_dat = pickle.loads(dat)
```

```python
type(my_dat)
```
