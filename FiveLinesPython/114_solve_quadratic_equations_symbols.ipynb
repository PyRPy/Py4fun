```python
# how to solve a quadratic equation in symbols
```

```python
from sympy import Symbol, solve, pprint
```

```python
x = Symbol('x')
```

```python
a = Symbol('a')
b = Symbol('b')
c = Symbol('c')
myEquation = a*x*x + b*x + c
```

```python
res = solve(myEquation, x, dict=True)
```

```python
type(res)
```

```python
pprint(res)
```
