```python
# what's zip in python, how to use it
# zip and enumerate
```

```python
student_id = [100, 101, 102, 103, 104]
student_name = ['David', 'Simon', 'Victor', 'Eric', 'Frank']
```

```python
students = zip(student_id, student_name)
```

```python
print(students)
```

```python
print(list(students))
```

```python
students = zip(student_id, student_name)
```

```python
for i, (student_id, student_name) in enumerate(students):
  print(i, student_id, student_name)
```
