```python
# define a class 'Dog'
# it has a name as well as color
```

```python
class Dog:
  def __init__(self, name, color):
    self.name = name
    self.color = color
  def bark(self):
    print("woof, woof!")
```

```python
# my dog's name is 'Lily', and she is white
lily = Dog("Lily", "White")
```

```python
print(hasattr(lily, 'color'))
```

```python
print(hasattr(lily, 'height'))
```
