```python
# python built-in functions: enumerate()
# enumerate : give index and value
```

```python
# my favorite list
fruits = ['apple', 'orange', 'banada']
```

```python
# print out a pair of index and value
for fruit in enumerate(fruits):
    print(fruit)
```

```python
# or you can do it this way using enumerate
for idx, fruit in enumerate(fruits):
    print(idx, fruit)
```

```python
# you can also change the starting index
for idx, fruit in enumerate(fruits, start = 10):
    print(idx, fruit)
```
