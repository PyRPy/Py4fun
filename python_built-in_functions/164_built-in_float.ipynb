```python
# python built-in functions: float()
```

```python
# convert integer to float
x = 4
print(float(x))
```

```python
# try it on a list
# it does not work!
myList = [1, 2, 3, 4, 5]
print(float(myList))
```

```python
# use lambda and map
myFloats = map(lambda x: float(x), myList)
print(list(myFloats))
```
