```python
# python built-in functions: compile()
```

```python
# define your code
myCode = 'x = 5\ny = 3\nz = x + y\nprint("x + y = ", z)'
```

```python
# compile before execution in python
codeToExec = compile(myCode, 'simple addition', 'exec')
```

```python
# run it 
exec(codeToExec)
```
