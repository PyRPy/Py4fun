```python
# python built-in functions: chr
```

```python
# guess what 97 is?
print(chr(97))
```

```python
nums = [97, 98, 99, 100, 101] 
```

```python
for num in nums:
    print(chr(num))
```

```python
# print letter from a to z 
for i in range(97, 123):
    print(chr(i))
```

```python
# use ord() to convert it back!
print(ord('a'))
```
