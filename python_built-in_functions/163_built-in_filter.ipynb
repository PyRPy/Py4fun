```python
# python built-in functions: filter()
```

```python
grades = [80, 50, 90, 100, 60]
```

```python
def grade_divider(y):
    if y < 60:
        return False
    else:
        return True
```

```python
performance = filter(grade_divider, grades)
```

```python
for res in performance:
    print(res)
```

```python
# number 50 is filtered out this time
```

```python
# you can also use lambda
performance2 = filter(lambda x: x >= 60, grades)
print(list(performance2))
```
