```python
# python built-in functions: format()
# format a list printout
```

```python
myList = [1.333, 12.25, 3.141516, 0.0, 2]
```

```python
f_list = ['{:.2f}'.format(num) for num in myList]
```

```python
print(f_list) # all strings in list
```

```python
# need to convert string to float in []
```

```python
f_list_float = map(lambda x: float(x), f_list)
```

```python
print(list(f_list_float))
```
