```python
# python built-in functions: exec()
```

```python
# write your code like you normal do
myCode = 'for i in range(5):\n\tprint(i)'

# '\n\t' just simulate new line and tab
```

```python
# compile before execution in python
codeToExec = compile(myCode, 
                     'simple loop', 
                     'exec')
```

```python
# run it 
exec(codeToExec)
```
