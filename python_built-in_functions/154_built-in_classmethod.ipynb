```python
# python built-in functions: classmethod()
```

```python
from datetime import date
```

```python
# let's build a dog class
# born year given, calculate age
class Dog:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  @classmethod  
  def bornYears(cls, name, year):
    return cls(name, date.today().year - year)
```

```python
# we have two dogs here
lily = Dog("Lily", 3)
jacky = Dog.bornYears("jacky", 2020)
```

```python
print("dog 'lily' age is: ", lily.age)
```

```python
# according to human
print("dog 'jacky' age is: ", jacky.age)
```
