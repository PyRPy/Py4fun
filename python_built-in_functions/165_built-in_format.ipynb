```python
# python built-in functions: format()
```

```python
# print result directly
num = 4.0 / 3.0
print('the result is: ', num)
```

```python
# print result with format you like
print('the result is {0:0.2f}'.format(num))
```

```python
# let's try more
num2 = 400000.0 / 3.0
```

```python
print('a large number: ', num2)
```

```python
print('a large number {0:.2f}'.format(num2))
```

```python
print('a large number {0:.2e}'.format(num2))
```
