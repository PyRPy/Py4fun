```python
class Dog:
  def __init__(self, name, color):
    self.name = name
    self.color = color
  def bark(self):
    print("woof, woof!")

lily = Dog("Lily", "White")
```

```python
print("my dog's name is ", lily.name)
print("my dog's color is ", lily.color)
```

```python
lily.bark()
```

```python
delattr(lily,'name')
```

```python
print(lily.name)
```

```python
print(lily.color)
```
