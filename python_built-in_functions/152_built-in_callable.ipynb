```python
# python built-in functions: callable
```

```python
def dummy():
    print('a very simple example')
test = dummy
```

```python
print(callable(test))
```

```python
# if it is a number
num = 9
print(callable(num))
```

```python
# build a class
class Dog:
    pass

dog = Dog()
```

```python
print(callable(dog))
```

```python
# add a call inside the class
class Dog:
    def __call__(self):
        print('call me a dog')
```

```python
dog = Dog()
print(callable(dog))
```
