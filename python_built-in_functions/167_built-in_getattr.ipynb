```python
# we have a simple class: dog
```

```python
class Dog:
  def __init__(self, name, color):
    self.name = name
    self.color = color
  def bark(self):
    print("woof, woof!")
```

```python
lily = Dog("Lily", "White")
```

```python
print(getattr(lily, 'color'))
```

```python
print(getattr(lily, 'weight'))
```
